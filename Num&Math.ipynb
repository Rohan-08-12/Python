```python
# Basic operations
10 + 5   # 15
10 - 5   # 5
10 * 5   # 50
10 / 5   # 2.0 (always float in Python 3)
10 // 3  # 3 (floor division - integer result)
10 % 3   # 1 (modulo - remainder)
10 ** 2  # 100 (exponent)

# Augmented assignment
x = 10
x += 5   # x = x + 5, now x = 15
x *= 2   # x = x * 2, now x = 30

# Math module
import math
math.sqrt(16)    # 4.0
math.ceil(4.3)   # 5
math.floor(4.7)  # 4
math.pi          # 3.141592...
math.sin(math.pi / 2)  # 1.0

# Random
import random
random.randint(1, 10)  # random integer between 1-10
random.choice(['a', 'b', 'c'])  # random element
random.random()  # float between 0-1
```
