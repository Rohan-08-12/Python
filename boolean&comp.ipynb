````python
# Comparison operators
5 == 5   # True (equality)
5 != 3   # True (not equal)
5 > 3    # True
5 <= 5   # True

# Logical operators (note: not &&, ||, !)
True and False  # False
True or False   # True
not True        # False

# Chaining comparisons (cool Python feature)
x = 5
1 < x < 10  # True (instead of: 1 < x and x < 10)

# Truthy/Falsy values
bool(0)       # False
bool("")      # False
bool(None)    # False
bool([])      # False
bool({})      # False

bool(1)       # True
bool("hi")    # True
bool([1])     # True

# Short-circuit evaluation
result = None or "default"  # "default"
result = "value" or "default"  # "value"
````

In JavaScript you use &&, ||, ! — in Python it's words:
&&     and    Both must be true
! not  
Flips True/False

## Falsy values (treated as False):
```python
bool(0)       # False (zero)
bool(0.0)     # False (zero float)
bool("")      # False (empty string)
bool(None)    # False (Python's null/undefined)
bool([])      # False (empty list)
bool({})      # False (empty dict)
bool(set())   # False (empty set)
```

## Truthy values (treated as True):
```python
bool(1)       # True (any non-zero number)
bool(-5)      # True
bool("hi")    # True (any non-empty string)
bool(" ")     # True (even just a space!)
bool([1])     # True (list with items)
bool([0])     # True (still has an item, even if it's 0!)
```
