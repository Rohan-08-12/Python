# Creation
person = {
    'name': 'Alex',
    'age': 25,
    'city': 'Toronto'
}

# Accessing
person['name']  # 'Alex'
person.get('age')  # 25
person.get('country', 'Canada')  # 'Canada' (default if key missing)

# Modifying
person['age'] = 26
person['job'] = 'Developer'  # add new key
del person['city']  # delete key

# Checking
'name' in person  # True
'country' in person  # False

# Iterating
for key in person:
    print(key, person[key])

for key, value in person.items():
    print(f"{key}: {value}")

for key in person.keys():
    print(key)

for value in person.values():
    print(value)

# Dict comprehension
squares = {x: x**2 for x in range(5)}
# {0: 0, 1: 1, 2: 4, 3: 9, 4: 16}
