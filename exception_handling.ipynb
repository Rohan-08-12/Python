```python
# WITHOUT exception handling (crashes):
number = int("abc")  # 💥 Program stops here
print("This never prints")

```

```python
# WITH exception handling (safe):
try:
    number = int("abc")
    print("Conversion successful")
except:
    print("Could not convert to number")  # This prints
print("Program continues")  # This also prints

# Output:
# Could not convert to number
# Program continues
```

```python
def divide_numbers(a_str, b_str):
    try:
        a = int(a_str)  # Might raise ValueError
        b = int(b_str)  # Might raise ValueError
        result = a / b  # Might raise ZeroDivisionError
        print(f"Result: {result}")
    
    except ValueError:
        print("Error: Please enter valid numbers")
    
    except ZeroDivisionError:
        print("Error: Cannot divide by zero")

# Test it:
divide_numbers("10", "2")    # Result: 5.0
divide_numbers("abc", "2")   # Error: Please enter valid numbers
divide_numbers("10", "0")    # Error: Cannot divide by zero
```

```python
try:
    number = int("abc")
except ValueError as e:
    print(f"Error occurred: {e}")

# Output:
# Error occurred: invalid literal for int() with base 10: 'abc'
```

```python
try:
    file = open('missing.txt', 'r')
except FileNotFoundError as e:
    print(f"File error: {e}")

# Output:
# File error: [Errno 2] No such file or directory: 'missing.txt'
```

```python
# The else Block
# Runs only if NO exception occurred:


try:
    number = int("42")
except ValueError:
    print("Invalid number")
else:
    print("Success! Number is valid")  # This runs
    print(f"Number: {number}")

# Output:
# Success! Number is valid
# Number: 42
```

```python
# The finally Block
# Runs ALWAYS - whether error happened or not:

try:
    file = open('data.txt', 'r')
    content = file.read()
except FileNotFoundError:
    print("File not found")
finally:
    print("This ALWAYS runs")
    # Clean up code goes here

# If file exists: reads it, then "This ALWAYS runs"
# If file missing: prints error, then "This ALWAYS runs"
```

```python
file = None
try:
    file = open('data.txt', 'r')
    content = file.read()
    print(content)
except FileNotFoundError:
    print("File not found")
finally:
    if file:
        file.close()  # Always close the file if it was opened
    print("Cleanup complete")
```

```python
# Raising Exceptions
# You can create your own errors using raise:

def set_age(age):
    if age < 0:
        raise ValueError("Age cannot be negative")
    if age > 150:
        raise ValueError("Age seems unrealistic")
    return age

try:
    my_age = set_age(-5)
except ValueError as e:
    print(f"Invalid age: {e}")

# Output:
# Invalid age: Age cannot be negative
```
