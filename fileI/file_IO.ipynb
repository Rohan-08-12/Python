```python
# ========================================
# WRITING FILES
# ========================================

# 1. Write a simple text file
with open('practice.txt', 'w') as file:
    file.write("Line 1\n")
    file.write("Line 2\n")
    file.write("Line 3\n")

print("✅ File created: practice.txt")

# ========================================
# READING FILES - Different Methods
# ========================================

# Method 1: Read entire file as one string
print("\n--- Method 1: read() ---")
with open('practice.txt', 'r') as file:
    content = file.read()
    print(content)

# Method 2: Read all lines as a list
print("\n--- Method 2: readlines() ---")
with open('practice.txt', 'r') as file:
    lines = file.readlines()
    print(lines)  # ['Line 1\n', 'Line 2\n', 'Line 3\n']

# Method 3: Loop through line by line (most common!)
print("\n--- Method 3: for loop ---")
with open('practice.txt', 'r') as file:
    for line in file:
        print(line.strip())  # .strip() removes \n

# ========================================
# APPENDING TO FILES
# ========================================

# Add more lines without erasing old content
with open('practice.txt', 'a') as file:
    file.write("Line 4 (appended)\n")
    file.write("Line 5 (appended)\n")

print("\n✅ Added more lines")

# Read again to see new content
with open('practice.txt', 'r') as file:
    print("\n--- File after append ---")
    print(file.read())

# ========================================
# WORKING WITH CSV-LIKE DATA
# ========================================

# Create a simple CSV
with open('students.txt', 'w') as file:
    file.write("Alice,25,A\n")
    file.write("Bob,30,B\n")
    file.write("Charlie,22,A\n")

# Read and parse it
print("\n--- Parsing CSV-like data ---")
with open('students.txt', 'r') as file:
    for line in file:
        name, age, grade = line.strip().split(',')
        print(f"{name} is {age} years old, grade: {grade}")

# ========================================
# COUNTING LINES, WORDS
# ========================================

# Create a longer file
with open('story.txt', 'w') as file:
    file.write("The quick brown fox\n")
    file.write("jumps over the lazy dog\n")
    file.write("Python is fun\n")

# Count lines
with open('story.txt', 'r') as file:
    lines = file.readlines()
    print(f"\n📊 Total lines: {len(lines)}")

# Count words
with open('story.txt', 'r') as file:
    content = file.read()
    words = content.split()
    print(f"📊 Total words: {len(words)}")

print("\n✅ All file operations complete!")
```

'r'   # Read (file must exist)
'w'   # Write (creates new file, ERASES old content!)
'a'   # Append (adds to end, doesn't erase)
'r+'  # Read and write
'w+'  # Write and read (erases old content)
