# Basic function
def greet(name):
    return f"Hello, {name}!"

result = greet("Alex")

# Default parameters
def power(base, exponent=2):
    return base ** exponent

power(5)     # 25 (uses default exponent=2)
power(5, 3)  # 125

# Keyword arguments
def describe_person(name, age, city):
    return f"{name}, {age}, from {city}"

describe_person("Alex", 25, "Toronto")
describe_person(age=25, city="Toronto", name="Alex")  # any order!

# *args (variable positional arguments)
def sum_all(*numbers):
    return sum(numbers)

sum_all(1, 2, 3)  # 6
sum_all(1, 2, 3, 4, 5)  # 15

# **kwargs (variable keyword arguments)
def print_info(**info):
    for key, value in info.items():
        print(f"{key}: {value}")

print_info(name="Alex", age=25, city="Toronto")

# Lambda (anonymous functions)
square = lambda x: x ** 2
square(5)  # 25

# Used with map, filter
numbers = [1, 2, 3, 4, 5]
squared = list(map(lambda x: x**2, numbers))
evens = list(filter(lambda x: x % 2 == 0, numbers))

# But list comprehensions are more Pythonic
squared = [x**2 for x in numbers]
evens = [x for x in numbers if x % 2 == 0]
