```python
import math

# Square root
math.sqrt(25)  # 5.0

# Power
math.pow(2, 3)  # 8.0 (same as 2**3)

# Rounding
math.ceil(4.2)   # 5 (round up)
math.floor(4.8)  # 4 (round down)

# Constants
math.pi  # 3.141592653589793
math.e   # 2.718281828459045

# Trigonometry
math.sin(math.pi / 2)  # 1.0
math.cos(0)            # 1.0

# Logarithms
math.log(10)      # Natural log
math.log10(100)   # 2.0 (log base 10)
```

```python
import random

# Random float between 0 and 1
print(random.random())  # 0.7234... (changes each time)

# Random integer in range
print(random.randint(1, 10))  # Random number 1-10 (inclusive)

# Random choice from list
fruits = ['apple', 'banana', 'cherry']
print(random.choice(fruits))  # Random fruit

# Shuffle a list (modifies in place)
numbers = [1, 2, 3, 4, 5]
random.shuffle(numbers)
print(numbers)  # [3, 1, 5, 2, 4] (random order)

# Random sample (pick multiple)
cards = ['A', 'K', 'Q', 'J', '10']
hand = random.sample(cards, 3)  # Pick 3 random cards
print(hand)  # ['Q', 'A', '10']
```

```python
from datetime import datetime, date, time

# Current date and time
now = datetime.now()
print(now)  # 2024-04-09 14:30:45.123456

# Just date
today = date.today()
print(today)  # 2024-04-09

# Extract parts
print(now.year)   # 2024
print(now.month)  # 4
print(now.day)    # 9
print(now.hour)   # 14

# Create specific date
birthday = date(1990, 5, 15)
print(birthday)  # 1990-05-15

# Date arithmetic
from datetime import timedelta
tomorrow = today + timedelta(days=1)
print(tomorrow)

# Format dates
print(now.strftime("%Y-%m-%d"))  # 2024-04-09
print(now.strftime("%B %d, %Y")) # April 09, 2024
```
