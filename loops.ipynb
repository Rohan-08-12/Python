```python
# list
fruits=["apple","banana","cherry"]

for f in fruits:
    print(f)
```

```python
prices=[10,20,30]
total=0

for p in prices:
    total+=p

print(f"Total : {total}")    
```

```python
# range() - Generate numbers
# range(stop) - 0 to stop-1
# range(start,stop) - start to stop-1
# range(start,stop,step) - start to stop-1, step

for i in range(5):
    print(i)
```

```python
for i in range(1,10):
    print(i)
```

```python
for i in range(1,10,2):
    print(i)
```

```python
for i in range(10,0,-1):
    print(i)
```

```python
# multiplication table
num=5
for i in range(1,11):
    print(f"{num} x {i} = {num*i}")
```

```python
# enumerate() - Get index and value

fruits=["apple","banana","cherry"]
for i,f in enumerate(fruits):
    print(f"Index:{i}, Fruit:{f}")
```

```python
# start from 1
fruits=["apple","banana","cherry"]
for i,f in enumerate(fruits,start=1):
    print(f"Index:{i}, Fruit:{f}")
```

```python
# zip() - Combine multiple lists

names=["Alice","Bob","Charlie"]
ages=[25,30,35]

for name,age in zip(names,ages):
    print(f"{name} is {age} years old")
```

```python
names=['Alice','Bob']
ages=[25,30]
cities=['NYC','LA']

for name,age,city in zip(names,ages,cities):
    print(f"{name} , {age} , from {city}")
```

```python
# While loop - Repeat until Condition is False

count=0

while count<5:
    print(count)
    count+=1
```

```python
# break

while True:
    answer = input("Type 'quit' to exit: ")
    if answer == 'quit':
        break  # Exit the loop
    print(f"You typed: {answer}")
```

```python
# continue -- Skip rest of iteration

for num in range(10):
    if num % 2 ==0:
        continue
    print(num)
```

```python
# Nested loop

for i in range(3):
    for j in range(2):
        print(f"({i} ,{j})")
```

```python
for i in range(1,4):
    for j in range(1,4):
        print(f"{i} x {j} = {i*j}")

    print("-----------")    
```
