```python
alice_hobbies = ['reading', 'gaming', 'cooking', 'reading', 'gaming']
bob_hobbies = ['gaming', 'sports', 'cooking', 'music']
```

```python
alice_hobbies=set(alice_hobbies)
bob_hobbies=set(bob_hobbies)
```

```python
common=alice_hobbies & bob_hobbies
print(common)
```

```python
diff=alice_hobbies - bob_hobbies
print(diff)
```

```python
unique= alice_hobbies | bob_hobbies
print(unique)
```
